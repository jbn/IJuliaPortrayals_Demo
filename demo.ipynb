You can write an `SVG` document inline, just like `HTML`. Or, you can include it `FromFile`.

```julia
SVG("""
<svg width="100" height="100">
  <circle cx="50" cy="50" r="40" stroke="green" stroke-width="4" fill="yellow" />
</svg>
""")
```

```julia
SVG(FromFile("ball.svg"))
```

JavaScript is useful if you're iteratively making nice [D3.js](http://d3js.org/) visualizations. Here, I just change some text.

```julia
HTML("""<div id="myel">No Text.</div>""")
```

```julia
JavaScript("""\$('#myel').text("Hello, World")""")
```

The `$`-interpolation is annoying sometimes. So, there is a JavaScript format string specifier.

```julia
JavaScript(JS"$('#myel').text('Hello, World -- again.')")
```
